```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load a sample dataset
df = sns.load_dataset("tips")  # Sample dataset with numeric & categorical data

# 1. Inspect the dataset
print("### Dataset Info ###")
df.info()
print("\n### Statistical Summary ###")
print(df.describe(include="all"))  # Include categorical summary too

# 2. Check for missing values and duplicates
print("\n### Missing Values ###")
print(df.isnull().sum())
print("\n### Duplicate Rows ###")
print(df.duplicated().sum())

# 3. Explore structure & relationships
print("\n### Number of Unique Values per Column ###")
print(df.nunique())

# Correlation matrix (numeric only)
print("\n### Correlation Matrix (Numeric Columns Only) ###")
print(df.select_dtypes(include="number").corr())

# 4. Identify trends & outliers with boxplot (numeric only)
plt.figure(figsize=(10,6))
df.select_dtypes(include="number").boxplot()
plt.title("Boxplot of Numerical Features")
plt.show()

# 5. Optional: visualize relationships with seaborn
sns.pairplot(df.select_dtypes(include="number"))
plt.show()

# 6. Optional: count plot for categorical columns
for col in df.select_dtypes(include="object").columns:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col)
    plt.title(f"Count plot of {col}")
    plt.show()
```
